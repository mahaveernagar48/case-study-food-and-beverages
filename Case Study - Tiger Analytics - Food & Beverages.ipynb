# Create a function to take out the flavors for each product

```python
def get_flavor(flavor):
    flavor_list = flavor.split(';')
    temp_list = []
    for flavor in flavor_list:
        flavor_1 = flavor.split(',')[0]
        if flavor_1 not in temp_list:
            temp_list.append(flavor_1)
    return temp_list
def get_sub_flavor(flavor):
    flavor_list = flavor.split(';')
    temp_list = []
    for flavor in flavor_list:
        flavor_1 = flavor.split(',')
        if len(flavor_1) == 1:
            temp_list.append(flavor_1[0])
        else:
            temp_list.append(flavor_1[1])
    return temp_list
product_launch_NA['Flavor_List'] = product_launch_NA['Flavor'].apply(lambda x: get_flavor(x))
product_launch_NA['Sub_Flavor_List'] = product_launch_NA['Flavor'].apply(lambda x: get_sub_flavor(x))
```

```python
product_launch_NA.head()
```

```python
product_launch_NA['Flavor Concat'] = product_launch_NA['Flavor'].apply(lambda x : str(x).replace(',','-') if ',' in str(x) else str(x))
def get_sub_flavor_1(flavor):
    flavor_list = flavor.split(';')
    temp_list = []
    for flavor in flavor_list:
        if flavor.lower().strip() not in temp_list:
            temp_list.append(flavor.lower().strip())
    return temp_list
product_launch_NA['Sub Flavor'] = product_launch_NA['Flavor Concat'].apply(lambda x: get_sub_flavor_1(x))
product_launch_NA.head()
```

# Analysing only last five years of data

```python
product_launch_NA_last5 = product_launch_NA[product_launch_NA['Launch Date'] > '2005-01-01']
product_launch_NA_last5.sort_values('Launch Date')
```

# Making all the flavors list

```python
temp = [x for x in product_launch_NA_last5['Flavor_List']]     
flavor_list = []
for i in temp:
    for j in i:
        if j not in temp:
            flavor_list.append(j.lower().strip())
flavor_list
```

# Counting each flavor

```python
from collections import Counter
flavor_Count = Counter(flavor_list)
flavor_Count
```

# Most common 10 flavors in last five years

```python
top_flavors = flavor_Count.most_common()
top_10_flavors = []
for i in top_flavors:
    if i[0].lower().strip() != 'not specified':
        top_10_flavors.append(i)
top_10_flavors = top_10_flavors[:10]
```

```python
flavor_dict = { i[0]:i[1] for i in top_10_flavors}
flavor_dict
```

```python
colors = ["red" if i == 'unflavored' else "skyblue" for i in list(flavor_dict.keys())]
plt.figure(figsize = (15,5), dpi = 100)
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
plt.bar(flavor_dict.keys(), flavor_dict.values(), color = colors)
addlabels(list(flavor_dict.keys()), list(flavor_dict.values()))
plt.xlabel("Flavors")
plt.ylabel("Flavor Count")
plt.xticks(rotation = 45)
plt.title("10 Emerging Flavors Over The Last 5 Years")
plt.show()
```

# Top 20 Emerging Sub Flavor Group Product in the last 5 years

```python
temp = [x for x in product_launch_NA_last5['Sub Flavor']]     
flavor_list = []
for i in temp:
    for j in i:
        if j not in temp:
            flavor_list.append(j.lower().strip())
flavor_Count = Counter(flavor_list)
top_flavors = flavor_Count.most_common()
top_10_flavors = []
for i in top_flavors:
    if i[0].lower().strip() != 'not specified':
        top_10_flavors.append(i)
top_10_flavors = top_10_flavors[:20]
flavor_dict = { i[0]:i[1] for i in top_10_flavors}
colors = ["red" if i == 'unflavored' else "skyblue" for i in list(flavor_dict.keys())]
plt.figure(figsize = (15,5), dpi = 100)
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
plt.bar(flavor_dict.keys(), flavor_dict.values(), color = colors)
addlabels(list(flavor_dict.keys()), list(flavor_dict.values()))
plt.xlabel("Sub Flavors Group")
plt.ylabel("Flavor Count")
plt.xticks(rotation = 90)
plt.title("20 Emerging Sub Flavors Group Over The Last 5 Years")
plt.show()
```

```python
markets = []
flavor = []
products = []
for market in flavor_dict:
    for key,value in flavor_dict[market].items():
        markets.append(market)
        flavor.append(key)
        products.append(value)
final_dict = {
    "Sub Market" : markets,
    "Flavors" : flavor,
    "Product Count":products
}
flavors_df = pd.DataFrame(final_dict)[0:25]
flavors_df['Concat Field'] = flavors_df['Sub Market'] +"-"+flavors_df['Flavors']
flavors_df
```

```python
plt.figure(figsize = (15,5), dpi = 100)
sns.barplot(x = 'Concat Field', y = 'Product Count', data = flavors_df, hue_order = 'Sub Market')
plt.xticks(rotation = 90)
plt.xlabel("Flavor Subcategories")
plt.show()
```
